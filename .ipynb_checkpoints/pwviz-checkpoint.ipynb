```python
import re
```

```python
newtext = ''
with open('./data/xom-all-flat-mod.xml', 'r') as pv:
    S = 0
    n = 0
    for line in pv.readlines():
        if re.search(r'<div\s+xml:lang', line):
            S = 1
            n = 0
        elif re.search(r'<p ', line) and S == 1:
            n += 1
            line = re.sub(r'<p[^>]*>', '<p n="{}">'.format(n), line)
        newtext += line
with open('./data/xom-all-flat-mod-pnums.xml', 'w') as pv:
    pv.write(newtext)
```
